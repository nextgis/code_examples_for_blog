```python
# GDAL - библиотека для работы с растровыми геоданными
from osgeo import gdal
# matplotlib - библиотека для визуализации данных
import matplotlib.pyplot as plt
```

```python
# Открываем COG растр из NGW прямой ссылкой
dataset = gdal.OpenEx('/vsicurl/https://demo.nextgis.com/api/resource/5519/cog')
```

```python
# Внутри - все прочитанные метаданные, объект типа gdal.Dataset
dataset
```

```python
# Количество каналов
dataset.RasterCount
```

```python
# Размеры растра
dataset.RasterXSize, dataset.RasterYSize
```

```python
%%time
# Напрямую прочитаем миллион пикселей и засечем время. Весь растр не будет грузиться - это особенность формата COG
data_array = dataset.ReadAsArray(0,0,1000,1000)
```

```python
# Самы значения пикселей
data_array
```

```python
# Их визуализация
plt.imshow(data_array)
```
